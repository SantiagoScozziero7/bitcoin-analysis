```python
import yfinance as yf
import pandas as pd
from datetime import date

# Descargar datos históricos de Bitcoin
btc = yf.download("BTC-USD", start="2015-01-01")

# Ver los primeros registros
print(btc.head())

# Guardar en la carpeta data
nombre = f"../data/bitcoin_{date.today()}.csv"
btc.to_csv(nombre)

print(f"\nArchivo guardado: {nombre}")
print(f"Total de filas: {len(btc)}")
```
