## Conclusión final del proyecto

A lo largo de este análisis utilizamos datos históricos de Bitcoin 
obtenidos de Yahoo Finance para responder las siguientes preguntas:

- ¿Cómo evolucionó el precio de Bitcoin a lo largo de los años?
- ¿Cuáles fueron los períodos de mayor volatilidad?
- ¿Cuál fue el rendimiento anual?
- ¿Qué relación existe entre volumen de operaciones y movimiento del precio?
- ¿En qué meses o años el mercado fue más estable o más volátil?

Gracias al análisis riguroso realizado sobre el dataset recopilado de 
Yahoo Finance, que se actualiza día a día, pudimos responder todas 
las preguntas satisfactoriamente.

**Evolución del precio:** El precio de Bitcoin tuvo un crecimiento 
extraordinario desde 2015, pasando de un promedio de $272 USD a 
$101.641 USD en 2025. El mayor incremento se registró en 2021, 
cuando el precio cuadruplicó su valor respecto al año anterior. 
Sin embargo, 2022 y 2023 mostraron caídas, reflejando la alta 
ciclicidad que caracteriza este mercado.

**Volatilidad:** Los años más volátiles fueron 2017, 2018 y 2021, 
todos con más del 4% de volatilidad diaria promedio, coincidiendo 
con los años de mayores movimientos de precio. Los más estables 
fueron 2016, 2023 y 2025 con volatilidades entre 2% y 2.5%.

**Rendimiento anual:** El mejor año fue 2017 con un rendimiento del 
1318%, mientras que los peores fueron 2018 y 2022 con caídas del 
-72% y -65% respectivamente. Esto confirma un patrón cíclico donde 
años de gran crecimiento son seguidos por fuertes correcciones.

**Volumen y precio:** La correlación entre volumen y rendimiento 
diario fue de -0.02, prácticamente nula, indicando que el volumen 
operado no es un predictor del movimiento del precio en el corto plazo.

**Estacionalidad:** Los meses más volátiles fueron marzo (3.70%), 
febrero (3.59%) y enero (3.46%), mientras que los más estables fueron 
octubre (2.32%), septiembre (2.76%) y agosto (3.07%), sugiriendo 
patrones estacionales que requerirían un análisis más profundo 
para ser confirmados.


## Glosario de términos

**Rendimiento diario:** Variación porcentual del precio de cierre de Bitcoin 
entre un día y el anterior. Indica cuánto subió o bajó el precio en un día.

**Volatilidad:** Medida que indica qué tan bruscos y dispares son los 
movimientos del precio día a día. Una volatilidad alta indica un mercado 
agitado, una volatilidad baja indica un mercado estable.

**Desviación estándar:** Medida estadística que indica qué tan dispersos 
están los valores respecto a su promedio. En este proyecto se usó para 
calcular la volatilidad.

**Rolling window (ventana móvil):** Técnica que consiste en calcular 
una métrica sobre un período fijo de días que va avanzando día a día. 
En este proyecto se usó una ventana de 30 días para calcular la volatilidad.

**Correlación:** Medida que indica qué tan relacionadas están dos variables. 
Va de -1 a 1, donde 1 significa relación perfecta positiva, 0 significa 
sin relación y -1 significa relación perfecta negativa.

**Volumen:** Cantidad total de dinero en dólares que se operó comprando 
y vendiendo Bitcoin en un día determinado.

**Crash:** Caída brusca y repentina del precio en un período muy corto 
de tiempo, generalmente provocada por pánico, incertidumbre o algún 
evento externo que sacude el mercado.

**Estabilidad:** Condición del mercado donde el precio se mueve poco 
día a día, reflejada en una volatilidad baja.

**Estacionalidad:** Patrón de comportamiento que se repite sistemáticamente 
en la misma época del año. En este proyecto se analizó si ciertos meses 
tienden a ser más volátiles que otros.

**Patrón cíclico:** Comportamiento que se repite a lo largo del tiempo. 
En Bitcoin se observó que años de gran crecimiento son seguidos 
por años de fuertes caídas.

**Mercado maduro:** Mercado con muchos participantes, alto volumen 
operado y menor volatilidad. Se contrapone a un mercado inmaduro 
donde hay pocos participantes y los movimientos de precio son 
más bruscos e impredecibles.

**Precio de cierre:** Último precio al que se operó Bitcoin al final 
del día. Es el precio más representativo y el más utilizado 
en análisis financiero.
